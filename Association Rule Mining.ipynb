```python
import pandas as pd
import io

filename = next(iter(uploaded))
data = pd.read_csv(io.BytesIO(uploaded[filename]))
print(data.head())
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules


# Convert values to binary
data = data.applymap(lambda x: 1 if x == 1 else 0)

support_count = 3
min_support = support_count / len(data)

# Run Apriori
frequent_apriori = apriori(data, min_support=min_support, use_colnames=True)
rules_apriori = association_rules(frequent_apriori, metric="confidence", min_threshold=0.5)

# Run FP-Growth
frequent_fpgrowth = fpgrowth(data, min_support=min_support, use_colnames=True)
rules_fpgrowth = association_rules(frequent_fpgrowth, metric="confidence", min_threshold=0.5)

# Compare rules
print("\n--- Apriori Rules ---")
print(rules_apriori[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

print("\n--- FP-Growth Rules ---")
print(rules_fpgrowth[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules

# Sample transaction dataset
transactions = [
    ['laptop', 'mouse', 'charger','USB','HDMI'],
    ['laptop', 'mouse','bag','USB','SSD'],
    ['charger', 'mouse','bag','USB','SSD'],
    ['mouse','USB','SSD'],
    ['laptop', 'mouse','charger','HDMI','bag'],
    ['laptop','USB','SSD'],
    ['laptop', 'mouse','charger','SSD'],
    ['mouse','USB','SSD','HDMI'],
]

# One-hot encode the data
te = TransactionEncoder()
te_data = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_data, columns=te.columns_)

# Find frequent itemsets using FP-Growth
frequent_itemsets = fpgrowth(df, min_support=0.3, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# Extract and print only the confidence values
confidence_df = rules[['antecedents', 'consequents', 'confidence']]
print(confidence_df)
```
